```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd

%matplotlib inline
```

## Упражнение 2

```python
x = np.linspace(-5, 5, 100)
y = x * x - x - 6

plt.figure(figsize = (9, 6))
plt.plot(x, y,
        linestyle = '-',
        linewidth = 2,
        color = 'blue',
        label = 'f(x)')
plt.title('График функции f(x)', fontsize = 17)

plt.xlabel('x', fontsize = 15)
plt.xlim(-3, 4)

plt.ylabel('f(x)', fontsize = 15)
plt.ylim(-8, 2)

plt.grid()

plt.show()
```

## Упражнение 3

```python
x = np.linspace(-50, 50, 10000)
y = np.log((x * x + 1) * np.exp(-np.absolute(x) / 10)) / np.log(1 + np.tan(1 / (1 + np.sin(x) * np.sin(x))))

plt.figure(figsize = (9, 6))
plt.plot(x, y,
        linestyle = '-',
        linewidth = 2,
        color = 'blue',
        label = 'f(x)')
plt.title('График функции f(x)', fontsize = 17)

plt.xlabel('x', fontsize = 15)
plt.xlim(-50, 50)

plt.ylabel('f(x)', fontsize = 15)
plt.ylim(0, 10)

plt.grid()

plt.show()
```
