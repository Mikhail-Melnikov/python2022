```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd

%matplotlib inline
```

## Упражнение 2

```python
x = np.linspace(-5, 5, 100)
y = x * x - x - 6

plt.figure(figsize = (9, 6))
plt.plot(x, y,
        linestyle = '-',
        linewidth = 2,
        color = 'blue',
        label = 'f(x)')
plt.title('График функции f(x)', fontsize = 17)

plt.xlabel('x', fontsize = 15)
plt.xlim(-3, 4)

plt.ylabel('f(x)', fontsize = 15)
plt.ylim(-8, 2)

plt.grid()

plt.show()
```

## Упражнение 3

```python
x = np.linspace(-50, 50, 10000)
y = np.log((x * x + 1) * np.exp(-np.absolute(x) / 10)) / np.log(1 + np.tan(1 / (1 + np.sin(x) * np.sin(x))))

plt.figure(figsize = (9, 6))
plt.plot(x, y,
        linestyle = '-',
        linewidth = 2,
        color = 'blue',
        label = 'f(x)')
plt.title('График функции f(x)', fontsize = 17)

plt.xlabel('x', fontsize = 15)
plt.xlim(-50, 50)

plt.ylabel('f(x)', fontsize = 15)
plt.ylim(0, 10)

plt.grid()

plt.show()
```

## Упражнение 5

```python
x = [1, 2, 3, 4, 5]
y = [0.99, 0.49, 0.35, 0.253, 0.18]

plt.figure(figsize = (9, 6))
plt.plot(x, y,
        linestyle = '-',
        linewidth = 2,
        color = 'blue',
        label = 'Исходные точки')
plt.title('График', fontsize = 17)

plt.xlabel('x', fontsize = 15)

plt.ylabel('f(x)', fontsize = 15)

plt.grid()
plt.errorbar(x, y, xerr = 0.05, yerr = 0.1)

x1 = np.linspace(0, 5, 1000)

p, v = np.polyfit(x, y, deg = 1, cov = True)
p_f = np.poly1d(p)
y1 = p_f(x1)
plt.plot(x1, y1,
        label = "Аппроксимация линейной функцией")

p2, v2 = np.polyfit(x, y, deg = 2, cov = True)
p_f2 = np.poly1d(p2)
y2 = p_f2(x1)
plt.plot(x1, y2,
        label = "Аппроксимация многочленом 2 степени")
plt.legend(title = '',
           fontsize = 12,
           title_fontsize = 13,
           shadow = True)

plt.show()
```

## Упражнение 6

```python
import math

xm = -2
xM = 2
x = np.linspace(xm, xM, 10000)
lim = 500
a = 3.0
b = 0.5
y = np.arange(len(x))

for i in range(len(x)):
    t = 0
    for n in range(lim):
        #t += b**n * math.cos(a**n * math.pi * x[i])
        v = a**n * x[i]
        o = v * math.pi
        #while o > 0:
        #    o -= 2 * math.pi * z
        t += b**n * math.cos(o % (2 * np.pi))
    y[i] = t

plt.figure(figsize = (9, 6))
plt.plot(x, y,
        linestyle = '-',
        linewidth = 2,
        color = 'blue',
        label = 'f(x)')
plt.title('График функции Вейерштрасса', fontsize = 17)

plt.xlabel('x', fontsize = 15)
plt.xlim(xm, xM)

plt.ylabel('f(x)', fontsize = 15)

plt.grid()

plt.show()
```
